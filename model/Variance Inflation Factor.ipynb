```python
import pandas as pd 
import numpy as np 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
data = pd.read_csv('finaldata.csv')
```

```python
X = data.drop(['url','type_airbnb','province','price_per_night','rating','rating_cleanliness','rating_accuracy','rating_communication','rating_location','rating_checkin','rating_value'],axis=1)
```

```python
X.head(1)
```

```python
X.shape
```

```python
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
```

```python
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
  
```

```python
pd.set_option('display.max_rows',None)
```

```python
vif_data
```

```python
vif_data[vif_data['VIF']!=np.inf].sort_values(by='VIF',ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
vif_data[vif_data['VIF']==np.inf].sort_values(by='VIF')
```
